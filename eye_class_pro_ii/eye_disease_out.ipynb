```python
from tensorflow.keras.models import load_model
import matplotlib.pyplot as plt
import numpy as np
import cv2

model_1=load_model('eye_own-model_1.3.h5')
img='E:\\RAM\\eye_desices\\OCT2017\\test\\DME\\DME-11053-1.jpeg'
```

```python
def shape_data(img):
    img=cv2.imread(img)
    img=cv2.resize(img,(128,128))
    img=cv2.normalize(img,None,alpha=0,beta=1,norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_32F)
    return img  
```

```python
cls=['CNV','NORMAL','DRUSEN','DME']

tt_img = shape_data(img)
tt_img1=np.array(tt_img).reshape(-1,128,128,3)
res=np.argmax(model_1.predict(tt_img1))
print(cls[res])
plt.imshow(tt_img)
```
